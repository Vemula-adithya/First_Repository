```python
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            text += page.extract_text()
    return text
```

```python
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    stemmer = PorterStemmer()
    tokens = word_tokenize(text.lower())
    filtered_tokens = [stemmer.stem(token) for token in tokens if token.isalnum() and token not in stop_words]
    return filtered_tokens
```

```python
def create_bag_of_words(tokens):
    freq_dist = FreqDist(tokens)
    bag_of_words = dict(freq_dist)
    return bag_of_words
```

```python
pdf_path = "ICF_002.pdf"
pdf_text = extract_text_from_pdf("ICF_002.pdf")
preprocessed_tokens = preprocess_text(pdf_text)
bag_of_words = create_bag_of_words(preprocessed_tokens)
print(len(bag_of_words))
```

```python
import PyPDF2

def split_and_merge_pdf(pdf_path, keywords, output_filename):
    pdf = PyPDF2.PdfReader(pdf_path)
    total_pages = len(pdf.pages)

    output_pdf = PyPDF2.PdfWriter()

    for page_number in range(total_pages):
        page = pdf.pages[page_number]
        text = page.extract_text()

        if any(keyword in text for keyword in keywords):
            output_pdf.add_page(page)

    with open(output_filename, "wb") as output_file:
        output_pdf.write(output_file)

    print(f"Merged PDF saved as {output_filename}")

# Example usage
pdf_path = "Consolidated_pdf.pdf"
output_filename = "Parental form_extracted_1.pdf"

split_and_merge_pdf(pdf_path, keywords, output_filename)
```

```python
import fitz  # PyMuPDF for PDF parsing
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

#Load the consolidated PDF and extract text from its pages
pdf_file = 'ICF_002.pdf'
pdf = fitz.open(pdf_file)

texts = []
for page_number in range(pdf.page_count):
    page = pdf.load_page(page_number)
    text = page.get_text()
    texts.append(text)

pdf.close()

#Prepare a labeled dataset
# Assuming a list of labels corresponding to each page in the consolidated PDF
labels = [0, 1, 0, 1, 2, 2, ...]

#Feature extraction using TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)

#Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

#Train the Naive Bayes classifier
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train, y_train)

#Predict the category of each page in the consolidated PDF
predicted_labels = nb_classifier.predict(X_test)

#Evaluate the classifier's performance if needed
#using metrics like accuracy_score, confusion_matrix, etc.

# Now you can use the classifier to predict the category of any new PDF page
new_text = "Text content of a new PDF page"
new_text_vectorized = vectorizer.transform([new_text])
predicted_category = nb_classifier.predict(new_text_vectorized)
```
